```python
import pandas as pd
import numpy as np
```

```python
jobs = jobs.replace(-1, np.nan)
jobs = jobs.replace("-1", np.nan)
jobs.head()
```

```python
jobs.info()
```

```python
jobs["Revenue"].unique()
```

```python
jobs["Revenue"].value_counts().plot.pie(figsize=(10, 8), autopct="%1.1f%%")
```

```python
jobs["Salary Estimate"].value_counts()
```

```python
jobs["Salary Estimate"].value_counts().plot.pie(figsize=(10, 8))
```

```python
jobs["Salary Estimate"].str.extract("(\d+)")
```

```python
jobs = jobs[jobs["Salary Estimate"].notnull()]
```

```python
jobs["Bottom Salary Estimate($K)"] = jobs["Salary Estimate"].str.extract("(\d+)").astype(int)
```

```python
jobs["Top Salary Estimte($K)"] = jobs["Salary Estimate"].str.extract("-\$(\d+)").astype(int)
```

```python
jobs.head()
```

```python
jobs["Bottom Salary Estimate($K)"].describe()
```

```python
jobs["Bottom Salary Estimate($K)"].max()
```

```python
pd.cut(jobs["Bottom Salary Estimate($K)"], np.arange(0, 120, 10))
```

```python
jobs["Top Salary Estimate($K)"].groupby(pd.cut(
    jobs["Top Salary Estimate($K)"],  np.arange(0, jobs["Top Salary Estimate($K)"].max() + 10, 10))).count().plot().ple(figsize=(10, 8))
```
